```python
#creating an histogram for the first feature
plt.figure(figsize=(11,6))
sns.histplot(x=df['fixed acidity'], color='orange', edgecolor='linen', alpha=0.5,bins=5)
plt.title('Histogram of Fixed Acidity')
plt.xlabel('fixed acidity', fontsize=20)
plt.ylabel('count', fontsize=20)
plt.show()
```

```python
#To cal stats for central tendancy
mean=round(df['fixed acidity'].mean(),2)
```

```python
print("Mean:",mean)
```

```python
median=df['fixed acidity'].median()
```

```python
print('Meadian:',median)
```

```python
mode=df['fixed acidity'].mode()
```

```python
print('Mode:',mode)
```

```python
plt.figure(figsize=(11,6))
sns.histplot(x=df['fixed acidity'], color='orange', edgecolor='linen', alpha=0.5,bins=5)
plt.title('Histogram of Fixed Acidity')
plt.xlabel('fixed acidity', fontsize=20)
plt.ylabel('count', fontsize=20)
plt.vlines(df['fixed acidity'].mean(),ymin=0,ymax=4000,colors='blue',label='mean')
plt.vlines(df['fixed acidity'].median(),ymin=0,ymax=4000,colors='red',label='median')
plt.legend()
plt.show()
```
